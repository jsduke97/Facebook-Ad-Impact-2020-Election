<h2>Experiment 1: Ad Impact on 2020 Presidential Election</h2>

```r
#-----------------------------------#
#----Read in data-------------------#
#-----------------------------------#
data_path = "./Data/"
```

```r
#----------------------------------------------------------------------------#
#----Biden ad spend by state from 2018-05-01 to 2020-11-03-------------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------biden_ad_dol----------------------------------------------------#
#--------Source: Meta Ad Library---------------------------------------------#
#----------------------------------------------------------------------------#
ads.biden <- read_excel(paste(data_path, "biden_ads_18-20_estadspendbyregion.xlsx", sep = ""))
ads.biden$state <- toupper(ads.biden$state)
#head(ads.biden)
```

```r
#----------------------------------------------------------------------------#
#----Trump ad spend by state from 2018-05-01 to 2020-11-03-------------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------trump_ad_dol----------------------------------------------------#
#--------Source: Meta Ad Library---------------------------------------------#
#----------------------------------------------------------------------------#
ads.trump <- read_excel(paste(data_path, "trump_ads_18-20_estadspendbyregion.xlsx", sep = ""))
ads.trump$state <- toupper(ads.trump$state)
#head(ads.trump)
```

```r
#----------------------------------------------------------------------------#
#----Votes for Trump/Biden by state in the 2020 general election-------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------candidate-------------------------------------------------------#
#------------votes-----------------------------------------------------------#
#--------Source: Harvard-----------------------------------------------------#
#----------------------------------------------------------------------------#
voting_2020 <- read_excel(paste(data_path, "harvard_dataverse_2020voting_clean.xlsx", sep = ""))
voting_2020_totals <- aggregate(votes ~ state, data = voting_2020, sum)
names(voting_2020_totals)[names(voting_2020_totals) == "votes"] <- "total_votes"
voting_2020 = merge(voting_2020, voting_2020_totals, by = "state")
#head(voting_2020)
```

```r
#----------------------------------------------------------------------------#
#----Voter registration details by state from 2020---------------------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------registered_voters - Total # of registered voters----------------#
#------------comb_dem - # of voters registered Democrat----------------------#
#------------comb_rep - # of voters registered Republican--------------------#
#--------Source: Independent Voter Project (28)------------------------------#
#----------------State Website (5)-------------------------------------------#
#----------------PEW Estimate (18)-------------------------------------------#
#----------------------------------------------------------------------------#
voter_reg <- read_excel(paste(data_path, "voter_reg_data.xlsx", sep = ""))
voter_reg = voter_reg[c('state','region','registered_voters', 'comb_dem', 'comb_rep', 'source')]

#head(voter_reg)
```

<h3>Merge data and print first few rows</h3>

```r
#----------------------------------------------------------------------------#
#----Merged Biden Ad Data----------------------------------------------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------biden_ad_dol - $ spent on facebook ads in the state-------------#
#------------undecided_won - votes in the state - registered democrats-------#
#----------------------------------------------------------------------------#
biden_merged <- merge(voting_2020[voting_2020["candidate"] == "JOSEPH R BIDEN",], ads.biden, by = "state")
biden_merged <- merge(biden_merged, voter_reg, by = "state")
biden_merged["vote_per"] <- biden_merged$votes / biden_merged$total_votes
biden_merged["reg_per"] <- biden_merged$comb_dem / (biden_merged$comb_dem + biden_merged$comb_rep)
biden_merged["undecided_won"] <- (biden_merged$vote_per - biden_merged$reg_per) * biden_merged$total_votes
biden_merged = biden_merged[c('state','region.x','biden_ad_dol', 'undecided_won')]
biden_merged = biden_merged[biden_merged$biden_ad_dol > 0,]
rownames(biden_merged) <- 1:nrow(biden_merged)

head(biden_merged)
```

```r
#----------------------------------------------------------------------------#
#----Merged Trump Ad Data----------------------------------------------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------trump_ad_dol - $ spent on facebook ads in the state-------------#
#------------undecided_won - votes in the state - registered republicans-----#
#----------------------------------------------------------------------------#
trump_merged <- merge(voting_2020[voting_2020["candidate"] == "DONALD J TRUMP",], ads.trump, by = "state")
trump_merged <- merge(trump_merged, voter_reg, by = "state")
trump_merged["vote_per"] <- trump_merged$votes / trump_merged$total_votes
trump_merged["reg_per"] <- trump_merged$comb_rep / (trump_merged$comb_rep + trump_merged$comb_dem)
trump_merged["undecided_won"] <- (trump_merged$vote_per - trump_merged$reg_per) * trump_merged$total_votes
#trump_merged[trump_merged$state == "MARYLAND",]
#trump_merged
trump_merged = trump_merged[c('state','region.x','trump_ad_dol', 'undecided_won')]
trump_merged = trump_merged[trump_merged$trump_ad_dol > 0,]
rownames(biden_merged) <- 1:nrow(biden_merged)

head(trump_merged)
```

<h3>Fit Models</h3>

```r
#----------------------------------------------------------------------------#
#----undecided_won ~ ad_spend + region---------------------------------------#
#----------------------------------------------------------------------------#
biden_merged.lm <- lm(undecided_won ~ biden_ad_dol + region.x, data = biden_merged)
```

```r
#----------------------------------------------------------------------------#
#----undecided_won ~ ad_spend + region---------------------------------------#
#----------------------------------------------------------------------------#
trump_merged.lm <- lm(undecided_won ~ trump_ad_dol + region.x, data = trump_merged)
```

```r
#----------------------------------------------------------------------------#
#----create log variables----------------------------------------------------#
#----------------------------------------------------------------------------#
#biden_merged$log_undecided_won <- log(biden_merged$undecided_won + 1)
biden_merged$log_biden_ad_dol <- log(biden_merged$biden_ad_dol)

#trump_merged$log_undecided_won <- log(trump_merged$undecided_won + 1)
trump_merged$log_trump_ad_dol <- log(trump_merged$trump_ad_dol)
```

```r
#----------------------------------------------------------------------------#
#----undecided_won ~ log(ad_spend) + region----------------------------------#
#----------------------------------------------------------------------------#
biden_merged.lm.linlog <- lm(undecided_won ~ log_biden_ad_dol + region.x, data = biden_merged)
trump_merged.lm.linlog <- lm(undecided_won ~ log_trump_ad_dol + region.x, data = trump_merged)

#summary(biden_merged.lm.linlog)
#summary(trump_merged.lm.linlog)
```

```r
cooksD <- cooks.distance(biden_merged.lm.linlog)
influential <- as.numeric(names(cooksD[cooksD > 5*mean(cooksD)]))
biden_merged.outliers <- biden_merged[influential,]
#biden_merged.outliers
biden_merged.nooutliers <- biden_merged[-influential,]
```

```r
cooksD <- cooks.distance(trump_merged.lm.linlog)
influential <- as.numeric(names(cooksD[cooksD > 5*mean(cooksD)]))
trump_merged.outliers <- trump_merged[influential,]
#trump_merged.outliers
trump_merged.nooutliers <- trump_merged[-influential,]
```

```r
#----------------------------------------------------------------------------#
#----undecided_won ~ log(ad_spend) + region no outliers----------------------#
#----------------------------------------------------------------------------#
biden_merged.lm.linlog2 <- lm(undecided_won ~ log_biden_ad_dol + region.x, data = biden_merged.nooutliers)
trump_merged.lm.linlog2 <- lm(undecided_won ~ log_trump_ad_dol + region.x, data = trump_merged.nooutliers)

#summary(biden_merged.lm.linlog2)
#summary(trump_merged.lm.linlog2)
```

```r
#----------------------------------------------------------------------------#
#----Biden summary-----------------------------------------------------------#
#----------------------------------------------------------------------------#
biden_merged.lm.rsq = round(summary(biden_merged.lm)$r.squared * 100, 4)
biden_merged.lm.adj_rsq = round(summary(biden_merged.lm)$adj.r.squared * 100, 4)
biden_merged.lm.coeff = round(summary(biden_merged.lm)$coef["biden_ad_dol",][1], 4)
biden_merged.lm.p_val = round(summary(biden_merged.lm)$coef["biden_ad_dol",][4], 4)

biden_merged.lm.linlog.rsq = round(summary(biden_merged.lm.linlog)$r.squared * 100, 4)
biden_merged.lm.linlog.adj_rsq = round(summary(biden_merged.lm.linlog)$adj.r.squared * 100, 4)
biden_merged.lm.linlog.coeff = round(summary(biden_merged.lm.linlog)$coef["log_biden_ad_dol",][1], 4)
biden_merged.lm.linlog.p_val = round(summary(biden_merged.lm.linlog)$coef["log_biden_ad_dol",][4], 4)

biden_merged.lm.linlog2.rsq = round(summary(biden_merged.lm.linlog2)$r.squared * 100, 4)
biden_merged.lm.linlog2.adj_rsq = round(summary(biden_merged.lm.linlog2)$adj.r.squared * 100, 4)
biden_merged.lm.linlog2.coeff = round(summary(biden_merged.lm.linlog2)$coef["log_biden_ad_dol",][1], 4)
biden_merged.lm.linlog2.p_val = round(summary(biden_merged.lm.linlog2)$coef["log_biden_ad_dol",][4], 4)



biden_merged.summary <- data.frame(c("R-squared", "Adj R-squared", "Ad P-Val", "Ad Coeff"),
            c(biden_merged.lm.rsq, biden_merged.lm.adj_rsq, biden_merged.lm.p_val, biden_merged.lm.coeff),
            c(biden_merged.lm.linlog.rsq, biden_merged.lm.linlog.adj_rsq, biden_merged.lm.linlog.p_val, biden_merged.lm.linlog.coeff),
            c(biden_merged.lm.linlog2.rsq, biden_merged.lm.linlog2.adj_rsq, biden_merged.lm.linlog2.p_val, biden_merged.lm.linlog2.coeff))
biden_merged.summary <- setNames(biden_merged.summary, 
                        c("", "lin-lin", "lin-log", "lin-log no outliers"))
biden_merged.summary
```

```r
#----------------------------------------------------------------------------#
#----Trump summary-----------------------------------------------------------#
#----------------------------------------------------------------------------#
trump_merged.lm.rsq = round(summary(trump_merged.lm)$r.squared * 100, 4)
trump_merged.lm.adj_rsq = round(summary(trump_merged.lm)$adj.r.squared * 100, 4)
trump_merged.lm.coeff = round(summary(trump_merged.lm)$coef["trump_ad_dol",][1], 4)
trump_merged.lm.p_val = round(summary(trump_merged.lm)$coef["trump_ad_dol",][4], 4)

trump_merged.lm.linlog.rsq = round(summary(trump_merged.lm.linlog)$r.squared * 100, 4)
trump_merged.lm.linlog.adj_rsq = round(summary(trump_merged.lm.linlog)$adj.r.squared * 100, 4)
trump_merged.lm.linlog.coeff = round(summary(trump_merged.lm.linlog)$coef["log_trump_ad_dol",][1], 4)
trump_merged.lm.linlog.p_val = round(summary(trump_merged.lm.linlog)$coef["log_trump_ad_dol",][4], 4)

trump_merged.lm.linlog2.rsq = round(summary(trump_merged.lm.linlog2)$r.squared * 100, 4)
trump_merged.lm.linlog2.adj_rsq = round(summary(trump_merged.lm.linlog2)$adj.r.squared * 100, 4)
trump_merged.lm.linlog2.coeff = round(summary(trump_merged.lm.linlog2)$coef["log_trump_ad_dol",][1], 4)
trump_merged.lm.linlog2.p_val = round(summary(trump_merged.lm.linlog2)$coef["log_trump_ad_dol",][4], 4)



trump_merged.summary <- data.frame(c("R-squared", "Adj R-squared", "Ad P-Val", "Ad Coeff"),
            c(trump_merged.lm.rsq, trump_merged.lm.adj_rsq, trump_merged.lm.p_val, trump_merged.lm.coeff),
            c(trump_merged.lm.linlog.rsq, trump_merged.lm.linlog.adj_rsq, trump_merged.lm.linlog.p_val, trump_merged.lm.linlog.coeff),
            c(trump_merged.lm.linlog2.rsq, trump_merged.lm.linlog2.adj_rsq, trump_merged.lm.linlog2.p_val, trump_merged.lm.linlog2.coeff))
trump_merged.summary <- setNames(trump_merged.summary, 
                        c("", "lin-lin", "lin-log", "lin-log no outliers"))
trump_merged.summary
```

```r
#----------------------------------------------------------------------------#
#----Registered voters and vote counts by region, gender 2016----------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------male_registered_16----------------------------------------------#
#------------female_registered_16--------------------------------------------#
#------------male_voted_16---------------------------------------------------#
#------------female_voted_16-------------------------------------------------#
#--------Source: US Census Bureau--------------------------------------------#
#----------------------------------------------------------------------------#
#voters_2016 <- read_excel(paste(data_path, "registered_voters_state_gender_2016.xlsx", sep = ""),range= cell_cols("A:F"))
voters_2016 <- read_excel(paste(data_path, "registered_voters_state_gender_2016.xlsx", sep = ""), sheet = "Sheet1", range= cell_cols("A:F"))
voters_2016 <- voters_2016[,c("State", "region", "male_voted_16", "female_voted_16")]

voters_2016.male = voters_2016[,c("State", "region", "male_voted_16")]
names(voters_2016.male)[names(voters_2016.male) == "male_voted_16"] <- "voted_16"
voters_2016.male$Gender = "MALE"
voters_2016.male = voters_2016.male[,c("State", "region", "Gender", "voted_16")]


voters_2016.female = voters_2016[,c("State","region", "female_voted_16")]
names(voters_2016.female)[names(voters_2016.female) == "female_voted_16"] <- "voted_16"
voters_2016.female$Gender = "FEMALE"
voters_2016.female = voters_2016.female[,c("State", "region", "Gender", "voted_16")]

voters_2016 = union(voters_2016.male, voters_2016.female)
voters_2016 = voters_2016[order(voters_2016$State, voters_2016$Gender),]
voters_2016$voted_16 = voters_2016$voted_16 * 1000

head(voters_2016)
```

```r
#----------------------------------------------------------------------------#
#----Registered voters and vote counts by region, gender 2020----------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------male_registered-------------------------------------------------#
#------------female_registered-----------------------------------------------#
#------------male_voted------------------------------------------------------#
#------------female_voted----------------------------------------------------#
#--------Source: US Census Bureau--------------------------------------------#
#----------------------------------------------------------------------------#
#voters_2020 <- read_excel(paste(data_path, "registered_voters_state_gender_2020.xlsx", sep = ""),range= cell_cols("A:F"))
voters_2020 <- read_excel(paste(data_path, "registered_voters_state_gender_2020.xlsx", sep = ""), sheet = "Sheet1", range= cell_cols("A:F"))

voters_2020.male = voters_2020[,c("State", "region", "male_voted")]
names(voters_2020.male)[names(voters_2020.male) == "male_voted"] <- "voted_20"
voters_2020.male$Gender = "MALE"
voters_2020.male = voters_2020.male[,c("State", "region", "Gender", "voted_20")]


voters_2020.female = voters_2020[,c("State","region", "female_voted")]
names(voters_2020.female)[names(voters_2020.female) == "female_voted"] <- "voted_20"
voters_2020.female$Gender = "FEMALE"
voters_2020.female = voters_2020.female[,c("State", "region", "Gender", "voted_20")]

voters_2020 = union(voters_2020.male, voters_2020.female)
voters_2020 = voters_2020[order(voters_2020$State, voters_2020$Gender),]
voters_2020$voted_20 = voters_2020$voted_20 * 1000

head(voters_2020)
```

```r
#----------------------------------------------------------------------------#
#----Merged Biden and Trump Ad Data 2016-2020--------------------------------#
#--------Columns:------------------------------------------------------------#
#------------state-----------------------------------------------------------#
#------------region----------------------------------------------------------#
#------------Unknown: Money spent on ads that reached unknown from 2018-2020-#
#------------Male: Money spent on ads that reached males from 2018-2020------#
#------------Female: Money spent on ads that reached females from 2018-2020--#
#------------male_voted: Count of males that voted in 2020 election----------#
#------------female_voted: Count of females that voted in 2020 election------#
#------------male_voted_16: Count of males that voted in 2016 election-------#
#------------female_voted_16: Count of females that voted in 2016 election---#
#----------------------------------------------------------------------------#
merged_adspend_2020 <- merge(ads.biden.gender, ads.biden.gender, by = c("State", "Gender"))
merged_adspend_2020$comb_ad_spend <- merged_adspend_2020$biden_ad_spend.x + merged_adspend_2020$biden_ad_spend.y
merged_adspend_2020 <- merged_adspend_2020[,c("State", "Gender", "comb_ad_spend")]
head(merged_adspend_2020)

merged_voted_gender_2020 <- merge(merged_adspend_2020,voters_2020,by=c('State', "Gender"),all=TRUE)
head(merged_voted_gender_2020)

merged_voted_gender_2016_2020 <-merge(merged_voted_gender_2020, voters_2016, by=c("State","region", "Gender"),all=TRUE)
head(merged_voted_gender_2016_2020)
```

```r
#----------------------------------------------------------------------------#
#------------Compute change in votes between 2016 and 2020 by gender---------#
#----------------------------------------------------------------------------#
merged_voted_gender_2016_2020$Difference_2016_2020 = merged_voted_gender_2016_2020$voted_20 - merged_voted_gender_2016_2020$voted_16
head(merged_voted_gender_2016_2020)
```

```r
#----------------------------------------------------------------------------#
#----Difference ~male$+male_voted+region-------------------------------------#
#----------------------------------------------------------------------------#
gender.lm <-lm(Difference_2016_2020 ~ comb_ad_spend + Gender + region, data = merged_voted_gender_2016_2020)
summary(gender.lm)


print('Affect of Ad money spent per region, gender- R-squared:39.86% Adjusted R-Squared:35.39% p-Value:1.25e-08')
```

```r
#----------------------------------------------------------------------------#
#----create log variables----------------------------------------------------#
#----------------------------------------------------------------------------#
merged_voted_gender_2016_2020$log_comb_ad_spend <- log(merged_voted_gender_2016_2020$comb_ad_spend + 1)
```

```r
#----------------------------------------------------------------------------#
#----Difference_2016_2020~ log(ad_spend) + gender + region-------------------#
#----------------------------------------------------------------------------#
merged_voted_gender_2016_2020.lm.linlog <- lm(Difference_2016_2020 ~ log_comb_ad_spend + Gender + region, data = merged_voted_gender_2016_2020)

summary(merged_voted_gender_2016_2020.lm.linlog)
```

```r
linlinrsq <- round(summary(gender.lm)$r.squared * 100, 2)
linlinarsq <- round(summary(gender.lm)$adj.r.squared * 100, 2)
linlincoeff = round(summary(gender.lm)$coef["comb_ad_spend",][1], 2)
linlinpval = round(summary(gender.lm)$coef["comb_ad_spend",][4], 6)
```
